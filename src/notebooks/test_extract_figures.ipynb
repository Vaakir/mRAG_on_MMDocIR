# Test Figure Extraction
Extract images, figures, charts, and tables from a single PDF using docling.

```python
import sys
from pathlib import Path

# Add src to path
SRC_DIR = Path("../").resolve()
sys.path.insert(0, str(SRC_DIR))

from preprocessing.extract_figures import _build_converter, extract_figures_from_pdf, VISUAL_LABELS
```

```python
# Paths
PDF_DIR = SRC_DIR / "data" / "train" / "pdf_train"
OUTPUT_DIR = SRC_DIR / "data" / "train" / "figures_train"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

# Pick a single PDF to test
pdf_files = sorted(PDF_DIR.glob("*.pdf"))
print(f"Found {len(pdf_files)} PDFs")
for i, p in enumerate(pdf_files[:10]):
    print(f"  [{i}] {p.name}")
```

```python
# Choose which PDF to test (change the index)
test_pdf = pdf_files[0]
print(f"Testing with: {test_pdf.name}")
```

```python
# Build the converter (loads models once)
converter = _build_converter(scale=2.0)
```

```python
# Extract figures from the test PDF
records = extract_figures_from_pdf(converter, test_pdf, OUTPUT_DIR)
print(f"Extracted {len(records)} images from {test_pdf.name}")
```

```python
# Show metadata for each extracted figure
for r in records:
    print(f"  {r['filename']}")
    print(f"    label={r['label']}  page={r['page_num']}/{r['total_pages']}  size={r['width']}x{r['height']}")
    print()
```

```python
# Display the extracted images with their metadata
from IPython.display import display
from PIL import Image

for r in records:
    img_path = OUTPUT_DIR / r["filename"]
    print(f"\n--- {r['label']} | page {r['page_num']}/{r['total_pages']} | {r['width']}x{r['height']}px ---")
    print(f"    {r['filename']}")
    img = Image.open(img_path)
    max_width = 600
    if img.width > max_width:
        ratio = max_width / img.width
        img = img.resize((max_width, int(img.height * ratio)))
    display(img)
```
