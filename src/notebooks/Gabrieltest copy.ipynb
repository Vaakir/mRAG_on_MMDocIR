## Example usage of LangChain:

```python
from langchain.agents import initialize_agent, AgentType, Tool
from langchain.llms import OpenAI

# Define tools the agent can use
tools = [
    Tool(name="SearchDocs", func=retrieve_documents, description="Search documents"),
    Tool(name="GenerateAnswer", func=generate_answer, description="Generate answer")
]

# Create agent
agent = initialize_agent(tools, OpenAI(), agent=AgentType.ZERO_SHOT_REACT_ACTION)

# Run
result = agent.run("What is the cash conversion cycle for Nike?")
```
